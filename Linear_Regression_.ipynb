```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
import math
```

**Linear Regression** uses the formula : Y = mX + C

where m = coefficient and C = intercept

Y is dependent variable and X is independent Variable

**Multi Varible Linear Regression:**

Y = m1*X1 + m2*X2 + m3*X3 + C

m1,m2,m3 are coefficients

X1,X2,X3 are features(independent variables)

C is intercept

Y is dependent variable





```python
df = pd.read_csv("/content/houseprice.csv")
df.head(3)
```

```python
mod1 = linear_model.LinearRegression()
mod1.fit(df[['area']],df.price)
```

```python
import pickle
```

```python
with open('model_pickle','wb') as f:
  pickle.dump(mod1,f)
```

```python
with open('model_pickle','rb') as f:
  mp = pickle.load(f)
```

```python
mp.predict([[5000]])
```

```python
plt.scatter(df.area,df.price,color='red',marker='+')
plt.plot(df.area,mod1.predict(df[['area']]),color='blue')
plt.show()
```

```python
df2 = pd.read_csv("/content/Book1.csv")
df2.head(3)
```

```python
p = mod1.predict(df2)
df2['price'] = p
df2.to_csv('output.csv')
```

```python
house_df = pd.read_csv("/content/Book2.csv")
house_df
```

```python
median_bedrooms = math.floor(house_df.bedrooms.median())
median_bedrooms
```

```python
house_df.bedrooms = house_df.bedrooms.fillna(median_bedrooms)
house_df
```

```python
mod2 = linear_model.LinearRegression()
mod2.fit(house_df[['area','bedrooms','age']],house_df.price)
```

```python
mod2.predict([[3000,4,15]])
```
